```python
df = pd.read_csv( r'C:\Users\kosma\Desktop\MAGISTER\FeatExtream\result_90500_temp.csv', index_col=None,  sep = ',', na_values = ['?'])
```

```python
def new_column(row):
    if math.isnan(row['label']):
        return row['label']
    else:
        return row['label'] > row['delay']

df['new_label'] = (df['label'] > df['delay']) & df['label']
df['new_label'] =  df.apply(new_column, axis=1)
```
